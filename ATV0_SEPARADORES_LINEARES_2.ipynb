### Dataset

```python
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
breast_cancer.keys()   
```

```python
X, y = breast_cancer.data, breast_cancer.target
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt
for i in range(3):
    plt.xlabel(f"{i}: {breast_cancer.feature_names[i]}")
    plt.ylabel("Classificação")
    plt.scatter(X[:,i], y)
    plt.show()
```

```python
X27 = X[:,27:28]
plt.xlabel(f"27: {breast_cancer.feature_names[27]}")
plt.ylabel("Classificação")
plt.scatter(X27, y)
plt.show()
     
```

### Classificador Linear

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

class ClassificadorLinear():
    def __init__(self):
        self.lrmodel = LinearRegression()
    def fit(self, X, y):
        self.lrmodel.fit(X,y)
    def predict(self, X):
        ypred = self.lrmodel.predict(X)
        ypred = (ypred>0.5).astype(int)
        return ypred

plt.xlabel(f"27: {breast_cancer.feature_names[27]}")
plt.ylabel("Classificação")
plt.scatter(X27, y)
modelo = ClassificadorLinear()
modelo.fit(X27,y)
ypred = modelo.predict(X27)
acuracia = accuracy_score(y, ypred)
plt.plot(X27, ypred, 'ro', label=f"acurácia: {acuracia}", alpha=0.5)
plt.legend()
plt.show()
```

```python
acuracia_1 = sum(y==ypred)/len(y)
acuracia_1
```

### Classificador Perceptron

```python
def sign(a):
  return (a >= 0) * 2 - 1
  
def accuracy(yactual, ypred):
  return sum(yactual == ypred)/len(yactual)
```

```python
from sklearn.base import BaseEstimator, ClassifierMixin
#import random
import numpy as np


def include_bias(X):
  bias = np.ones((X.shape[0], 1))
  Xb = np.concatenate((bias, X), axis=1)
  return Xb

class Perceptron(BaseEstimator, ClassifierMixin):
  def __init__(self, max_iter=100):
     self.w = None
     self.activation = sign
     self.max_iter = max_iter

  def fit(self, X, y):
    Xb = include_bias(X)
    self.w = np.random.uniform(-1, 1, size=Xb.shape[1])
    for _ in range(self.max_iter):
      ypred = self.predict(X)
      erros = y - ypred
      custo = sum(erros**2)
      if custo == 0:
        break
      self.w += Xb.T @ erros

  def predict(self, X):
    Xb = include_bias(X)
    a = Xb @ self.w
    ypred = self.activation(a)
    return ypred

Xtr = X27
ytr = y

perceptron = Perceptron()
perceptron.fit(Xtr, ytr)
print(accuracy(ytr, perceptron.predict(Xtr)))
print(perceptron.w)
```

```python

plt.xlabel(f"27: {breast_cancer.feature_names[27]}")
plt.ylabel("Classificação")
plt.scatter(X27, y)
perceptron = Perceptron()
perceptron.fit(X27,y)
ypred = perceptron.predict(X27)
acuracia_2 = accuracy_score(y, ypred)
plt.plot(X27, ypred, 'ro', label=f"acurácia: {acuracia_2}", alpha=0.5)
plt.legend()
plt.show()
```

### Classificador ScikitLearn

```python
plt.xlabel(f"27: {breast_cancer.feature_names[27]}")
plt.ylabel("Classificação")
plt.scatter(X27, y)
modelo = LinearRegression()
modelo.fit(X27,y)
ypred = modelo.predict(X27)
ypred = (ypred>0.5).astype(int)
acuracia_3 = accuracy_score(y, ypred)
plt.plot(X27, ypred, 'ro', label=f"acurácia: {acuracia_3}", alpha=0.5)
plt.legend()
plt.show()
```
